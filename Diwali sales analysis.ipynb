### **Exploratory data analysis**

Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df, color='skyblue')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, color='skyblue')
```

Based on the two graphs analyzing sales by gender, it's clear that the majority of purchases were made by women. The countplot shows a higher number of transactions initiated by women, and the barplot of total sales amount confirms that women contributed significantly more to the overall revenue compared to men. This suggests that women are a key demographic for the business, and future marketing efforts might be most impactful if they primarily target this group.

### **Age**

```python
sns.countplot(data = df, x = 'Age Group', hue='Gender', palette='viridis')
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, color='skyblue')
```

from above graphs we can see that most of the buyers are of age group between 26-35 yrs


```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders', color='skyblue')
```

From above graph we can see that most of the orders and total sales/amount from uttar pradesh, maharashtra, and karnataka respectively.

**Marital status**

```python
ax = sns.countplot(data = df, x = 'Marital_Status', color='skyblue')
sns.set(rc={'figure.figsize':(4,2)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender', palette='viridis')
```

from above graphs we can see that most of the buyers are married(women) and they have high purchasing power

# **Occupation**

```python
sns.set(rc={'figure.figsize':(20,5)})
```

```python
ax = sns.countplot(data = df, x = 'Occupation', palette='viridis')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount', palette='viridis')
```

From above graphs we can see that most of the buyers are working in IT,Aviation and Healthcare sector
